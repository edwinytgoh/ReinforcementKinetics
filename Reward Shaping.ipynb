```python
%matplotlib notebook 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from envs.SimUtils import sigmoid as sigm
```

```python
x = np.linspace(-10,100, 10000)/100
```

```python
0.9*0.5
```

# Temp Reward

```python
y = 100*sigm(-x)
y2 = 100*sigm(-(x-10))
y3 = 100*sigm(-2*(x-10))
y4 = 100*sigm(-150*(x-0.025))
y5 = 100*sigm(-250*(x-0.025))
y6 = 100*sigm(-450*(x-0.045))
plt.subplots()
plt.plot(x, y, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')
plt.plot(x, y4, label='y4')
plt.plot(x, y5, label='y5')
plt.plot(x, y6, label='y6')
plt.grid()
plt.xlim((0, 0.06))
plt.legend()
```

# CO Reward

```python
plt.subplots()
x = np.linspace(0, 1000, 10000)
y = 100*sigm(-3*(x-30))
y2 = 100*sigm(-3*(x-32))

plt.plot(x, y, label='y1')
plt.plot(x, y2, label='y2')
plt.grid()
plt.xlim((20, 35))
plt.legend()
```

# NO Reward

```python
plt.subplots()
x = np.linspace(0, 1000, 10000)
y = 200*sigm(-(x-25))
y2 = 200*sigm(-0.4*(x-15))

plt.plot(x, y, label='y1')
plt.plot(x, y2, label='y2')
plt.grid()
plt.xlim((0, 35))
plt.legend()
```

# Reactant Reward

```python
plt.subplots() 
x = np.linspace(0,100,1000)
y1 = (100 - x)
y2 = np.square(y1)
y3 = np.exp(y1)
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')
plt.yscale('log')
plt.grid()
```
